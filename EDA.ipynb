En el caso particular de la ciudad de Buenos Aires, el EDA nos permitirá desentrañar las complejas dinámicas subyacentes a los siniestros viales. A través de la aplicación de técnicas estadísticas descriptivas y visualizaciones de datos, se busca identificar patrones, tendencias y correlaciones entre diversas variables, tales como condiciones climáticas, tipo de vehículo, horario del incidente y características de la vía. Este proceso investigativo no solo permitirá comprender la magnitud y naturaleza del problema, sino que también sentará las bases para la formulación de hipótesis y el diseño de modelos predictivos más sofisticados.
